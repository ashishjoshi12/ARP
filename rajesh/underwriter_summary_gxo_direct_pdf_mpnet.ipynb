```python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

# Load PDF
pdf_path = "/content/drive/My Drive/ARM/gxo_10k.pdf"
loader = PyPDFLoader(pdf_path)
documents = loader.load()

# Split text into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
docs = text_splitter.split_documents(documents)

# Generate embeddings using MPNet
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")

# Create **In-Memory** FAISS vector store
vector_db = FAISS.from_documents(docs, embedding_model)

print("✅ 10-K PDF processed and stored in FAISS (In-Memory)!")
```

```python
from langchain.chains import RetrievalQA
from langchain_ollama import ChatOllama  # ✅ Updated import

# Load Ollama (Llama3)
llm = ChatOllama(model="llama3")

# Create Retrieval Chain (corrected)
qa_chain = RetrievalQA.from_chain_type(llm=llm, retriever=vector_db.as_retriever())

# Ask a financial question
# query = "What are the key reasons for revenue increase in 2023?"

# query = "What factors influenced the revenue trend in 2023?"


questions = [
    "What factors influenced the revenue trend in 2023?",
"""
What factors influenced the transaction and integration costs trend in 2023?
"""
]


# response = qa_chain.invoke({"query": query})  # ✅ Use `invoke` instead of `run`


# print("📌 Answer:", response['result'])

for question in questions:
    print(f"Question: {question}")
    response = qa_chain.invoke({"query": question})
    print("📌 Answer:", response['result'])
    print("-------------------------------------------------------------------")
```
