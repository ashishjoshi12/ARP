```python
import pandas as pd

# Load the financial statement
#file_path = '/content/drive/My Drive/ARM/financial_statement.xlsx'
file_path = '/content/drive/My Drive/ARM/gxo_fin_stmt.xlsx'
df = pd.read_excel(file_path)
col_name = 'Category'

# Normalize column names (strip spaces, lowercase for case-insensitivity)
df[col_name] = df[col_name].str.strip().str.lower()

def get_financial_row(param):
    """Function to filter a row based on the given parameter."""
    param = param.strip().lower()  # Normalize input

    # Define alias mappings for common terms
    alias_mapping = {
        "cogs": "cost of goods sold (cogs)",
        "revenue": "revenue"
    }

    # Standardize parameter
    search_term = alias_mapping.get(param, param)

    # Filter the row
    filtered_row = df[df[col_name] == search_term]

    return filtered_row if not filtered_row.empty else "No matching record found."

# Example usage
param = "Revenue"  # Change this to "cogs" or other terms
financial_data = get_financial_row(param)
print(financial_data)
```

```python
import re

# Define the function to check if a chunk contains revenue-related content
def is_revenue_related(chunk):
    # Check for keywords like "Revenue", "Sales", "Income", etc.
    revenue_keywords = ["Revenue", "Sales", "Total Income", "Profit", "Growth"]
    for keyword in revenue_keywords:
        if re.search(r'\b' + re.escape(keyword) + r'\b', chunk, re.IGNORECASE):  # Case-insensitive match
            return True
    return False

# Assuming `chunks` is your list containing all 2035 chunks
# Count the number of revenue-related chunks
revenue_count = sum(1 for chunk in chunks if is_revenue_related(chunk))

# Print the total number of revenue-related chunks out of the total number of chunks
print(f"Total revenue-related chunks: {revenue_count} out of {len(chunks)}")
```

```python
from langchain_ollama import OllamaEmbeddings
import chromadb
import time
import re

# Define batch size
BATCH_SIZE = 5  # Process 5 at a time to avoid memory issues

# Initialize embedding model
embedding_model = OllamaEmbeddings(model="llama3")

# Initialize ChromaDB with persistent storage
client = chromadb.PersistentClient(path="./chroma_db")
collection_name = "financial_reports_10k5"

# Try to get the collection, and create it if it doesn't exist
collection = None
try:
    collection = client.get_collection(collection_name)
    print(f"Collection '{collection_name}' found.")
except Exception:
    collection = client.create_collection(collection_name)
    print(f"Collection '{collection_name}' created.")

# Function to check if chunk contains revenue-related content
def is_revenue_related(chunk):
    # Check for keywords like "Revenue", "Sales", "Income", etc.
    revenue_keywords = ["Revenue", "Sales", "Total Income", "Profit", "Growth"]
    for keyword in revenue_keywords:
        if re.search(r'\b' + re.escape(keyword) + r'\b', chunk, re.IGNORECASE):
            return True
    return False

# Debug: Check total chunks
print(f"Total chunks available: {len(chunks)}")

# Process chunks **sequentially**
for start_idx in range(0, len(chunks), BATCH_SIZE):
    batch_chunks = chunks[start_idx:start_idx + BATCH_SIZE]

    print(f"\nProcessing batch {start_idx // BATCH_SIZE + 1} with {len(batch_chunks)} chunks...")

    try:
        # Prepare metadata and filtered chunks for revenue-related content
        metadatas = []
        filtered_chunks = []  # To store only revenue-related chunks
        filtered_embeddings = []  # To store embeddings for filtered chunks

        # Check for revenue-related content and generate embeddings only for those chunks
        for j, chunk in enumerate(batch_chunks):
            # Check if the chunk contains revenue-related content
            if is_revenue_related(chunk):
                # Generate embedding only for revenue-related chunks
                batch_embedding = embedding_model.embed_documents([chunk])[0]  # Generate embedding for the chunk

                # Add metadata and store filtered chunks and embeddings
                metadatas.append({
                    "source": "10K",
                    "section": f"chunk_{start_idx+j}",
                    "year": "2023",
                    "revenue": True  # Mark this chunk as revenue-related
                })
                filtered_chunks.append(chunk)  # Add to the list of revenue-related chunks
                filtered_embeddings.append(batch_embedding)  # Add corresponding embedding

        # Store only revenue-related chunks and embeddings in ChromaDB
        if filtered_chunks:  # Proceed only if there are revenue-related chunks
            collection.add(
                documents=filtered_chunks,  # Only revenue-related chunks
                metadatas=metadatas,  # Metadata for revenue-related chunks
                ids=[f"chunk_{start_idx+j}" for j in range(len(filtered_chunks))],  # Unique ids for each chunk
                embeddings=filtered_embeddings  # Only embeddings for revenue-related chunks
            )

            print(f"✅ Stored revenue-related batch {start_idx // BATCH_SIZE + 1} in ChromaDB. Total records: {collection.count()}")

        else:
            print(f"No revenue-related chunks in batch {start_idx // BATCH_SIZE + 1}.")

    except Exception as e:
        print(f"❌ Error processing batch {start_idx // BATCH_SIZE + 1}: {e}")

print("\n🚀 All revenue-related chunks stored successfully in ChromaDB!")
```

```python
query = """
Provide a breakdown of revenue-related sections from the 10-K financial report:
1. Revenue amount for each section.
2. Reasons for any increase or decrease in revenue.
3. Additional insights on revenue growth or decline.
"""

# Convert query into embedding
query_embedding = embedding_model.embed_query(query)

# Perform similarity search
results = collection.query(query_embeddings=[query_embedding])

# Extract relevant text chunks
if results.get('documents'):
    combined_documents = "\n\n".join(map(str, results['documents']))  # Convert to text

    # Construct prompt for Llama3
    prompt = f"""
    Please analyze the following financial data and provide insights on:
    1. Revenue amount per section.
    2. Reasons for revenue increase or decrease.
    3. Additional insights into revenue growth or decline.

    Documents:
    {combined_documents}
    """

    try:
        # Send the extracted data to Llama3 for analysis
        response = ollama.chat(model="llama3", messages=[
            {"role": "system", "content": "You are a financial analyst providing revenue insights."},
            {"role": "user", "content": prompt}
        ])

        # Display response
        print("\n💡 **Revenue Analysis:**\n")
        print(response.get('message', {}).get('content', 'No response from model'))

    except Exception as e:
        print(f"❌ Error processing the documents: {e}")

else:
    print("⚠️ No relevant revenue details found in the database.")
```
